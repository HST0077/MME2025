```python
def sieve_of_eratosthenes(n):
    # 0 n  True  
    is_prime = [True] * (n + 1)

    # ,  False 
    is_prime[0] = is_prime[1] = False

    #   n  
    for i in range(2, int(n**0.5) + 1):
        if is_prime[i]: # i True
            # i*(i),i*(1+1),...  False 
            for j in range(i * i, n + 1, i):
                is_prime[j] = False
    # is_prime index(num),(prime)  
    # prime True index 
    primes = [num for num, prime in enumerate(is_prime) if prime]
    return primes #  index 
```

```python
#  
n = 10000
timer.tic()
result = sieve_of_eratosthenes(n)
print(f"{n}   : {len(result)}")

timer.toc()
```

```python
timer.tic()
result = sieve_of_eratosthenes(1000000)
print(f"{n}   : {len(result)}")
timer.toc()
```

```python
# Recursive Python function to solve the tower of hanoi
Min_Move=0
def TowerOfHanoi(n , source, destination, auxiliary):
    if n==1: #  n=1  .
        print (f'1   {source} {destination} ')
        return #       .
    #  n>=2 n-1  2   
    TowerOfHanoi(n-1, source, auxiliary, destination)
    # n  3 .
    print (f'{n}   {source} {destination} ')
    # 2   n-1   3  
    TowerOfHanoi(n-1, auxiliary, destination, source)
```

```python
#    
n = 3
TowerOfHanoi(n,'1','3','2')
```

```python
#     .
def HTowerMove(n):
    if n==1: # n=1
        return 1 #  1 return.
    else: # n>=2
        # n-1   2 1   .
        return 2*HTowerMove(n-1)+1
HTowerMove(4)
```

```python
# shift() 
Test_df=NYA_df[["Adj Close"]].head(5) #     5 
Test_df['lag_1']=Test_df['Adj Close'].shift(1) # 1 
Test_df
```

```python
# shift() 
D=NYA_df["Adj Close"].head(5) #     5 
Test_df=pd.DataFrame(data={'Price':D}) #  ,  Price 
Test_df['lag_1']=Test_df.Price.shift(1) # 1 
Test_df['lag_2']=Test_df.Price.shift(2) # 2 
Test_df['lag_-1']=Test_df.Price.shift(-1) # 1  
Test_df
```

```python
# (Moving Average) 
Test_df['MA']=Test_df.Price.rolling(3).mean() # 3  .
Test_df.head(10) #  10 
```
