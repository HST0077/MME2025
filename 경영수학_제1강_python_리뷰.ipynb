```python
# 피보나치 수열을 생성함과 동시에 부분합의 크기를 체크하면서 10000이 넘는 순간 중단 시킨다.
import numpy as np
F=np.zeros(10000) # 피보나치 수열 10000개를 담기 위한 numpy array를 만듭니다.
F[0],F[1]=0,1 # F[0]=0, F[1]=1 값을 초기값으로 저장
Ans=F[0]+F[1]
for i in range(2,10001): # 2부터 시작해서 10000까지 1씩 증가시킨다.
    F[i]=F[i-1]+F[i-2] # 피보나치 수열의 정의: 직전 2개의 항의 합으로 현재의 항을 생성
    Ans+=F[i] # i번째 항까지의 부분합을 구한다.
    if Ans > 10000: # 부분합이 10000을 넘는다면,
        print(f'{i+1}번째 수열까지의 합이 처음으로 10000을 넘었습니다. 그 값은 {Ans}입니다.')
        break # for 문을 exit 시킨다.
```

```python
from pytictoc import TicToc
import numpy as np
timer=TicToc() # 시간측정 객체 만들기
timer.tic() # 측정 시작
n=1000
G_pot=[] # 소수 리스트
for i in range(2,n+1):
    i_pot=[] # i의 약수 리스트
    for j in range(1,i+1):
        if np.mod(i,j)==0: #j는 i의 약수
            i_pot.append(j)
    if len(i_pot) == 2:
        G_pot.append(i)
print(f"{n} 이하의 소수의 개수: {len(G_pot)}")
timer.toc() # 측정후 시간 출력
```

```python
def sieve_of_eratosthenes(n):
    # 0부터 n까지 모두 True인 리스트 생성
    is_prime = [True] * (n + 1)

    # 첫번째, 두번째는 False로 설정
    is_prime[0] = is_prime[1] = False

    # 세번째 부터 n의 제곱근인 자연수까지
    for i in range(2, int(n**0.5) + 1):
        if is_prime[i]: # i번째가 True이면
            # i*(i),i*(1+1),...를 모두 False로 설정
            for j in range(i * i, n + 1, i):
                is_prime[j] = False
    # is_prime의 index(num),값(prime)을 하나씩 빼서
    # prime이 True면 index만 추출
    primes = [num for num, prime in enumerate(is_prime) if prime]
    return primes # 총 index를 리턴
```

```python
# 사용 예시
n = 10000
timer.tic()
result = sieve_of_eratosthenes(n)
print(f"{n} 이하의 소수의 개수: {len(result)}")

timer.toc()
```

```python
timer.tic()
result = sieve_of_eratosthenes(1000000)
print(f"{n} 이하의 소수의 개수: {len(result)}")
timer.toc()
```

```python
# Recursive Python function to solve the tower of hanoi
Min_Move=0
def TowerOfHanoi(n , source, destination, auxiliary):
    if n==1: # 만일 n=1이면 한번에 끝난다.
        print (f'1 번째 원반을 {source}에서 {destination}으로 옮긴다')
        return # 리턴 값이 없으면 그냥 공란으로 두면 된다.
    # 만일 n>=2이면 n-1개를 먼저 2번기둥으로 옮기는 문제를 풀고
    TowerOfHanoi(n-1, source, auxiliary, destination)
    # n번째 원반을 3번기둥으로 옮긴다.
    print (f'{n} 번째 원반을 {source}에서 {destination}으로 옮긴다')
    # 2번 기둥에 있던 n-1개의 원반을 이제 3번 기둥으로 옮긴다
    TowerOfHanoi(n-1, auxiliary, destination, source)
```

```python
# 최소 움직임 회수만 계산해 보자.
def HTowerMove(n):
    if n==1: # n=1이면
        return 1 # 답 1을 return한다.
    else: # n>=2이면
        # n-1일 때 회수의 2배에 1을 더해 주면 된다.
        return 2*HTowerMove(n-1)+1
HTowerMove(4)
```

```python
# 인덱스와 칼럼 이름 생성
C=[f'c{i}' for i in range(1,11)] # c1, ... c10 생성
R=[f'r{i}' for i in range(1,11)] # r1, ... r10 생성

# 데이터 생성
D=np.array([i for i in range(100)]) # 0,1,2,...,99 생성
D=D.reshape(10,10) # 10 by 10으로 재배치 시킴

# 데이터프레임으로 변환
df=pd.DataFrame(data=D,columns=C,index=R)
df
```

```python
# 칼럼합 구하기
# f문을 이용해 dictionary 형태 자료 생성
# {'c2':'C_Grp','c4':'C_Grp','c6':'C_Grp','c8':'C_Grp','c10':'C_Grp'}
dic={f'c{2*i}':'C_Grp' for i in range(1,6)}
df.groupby(dic,axis=1).sum() # axis=0은 행 방향, axis=1은 칼럼방향이 됨
```

```python
# row들에 대한 부분합 구하기
# f문을 이용해 dictionary 형태 자료 생성
# {'r1': 'R_Grp', 'r3': 'R_Grp', 'r5': 'R_Grp', 'r7': 'R_Grp', 'r9': 'R_Grp'}
dic={f'r{2*i-1}':'R_Grp' for i in range(1,6)}
df.groupby(dic,axis=0).sum() # axis=0은 생략해도 된다.
```

```python
# 사용자 정의 함수를 만들어서
R_sum=[f'r{2*i-1}' for i in range(1,6)] # r1,r3,r5,r7,r9
def row_sum(x):
    if x in R_sum: # x가 R_sum에 속한다면
        return 'R_Grp'

df.groupby(row_sum,axis=0).sum()
```
